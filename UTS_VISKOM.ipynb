```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
```

```
from google.colab import files
uploaded = files.upload()

# Ambil nama file
image_path = list(uploaded.keys())[0]

# Baca gambar
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Citra Asli")
plt.axis('off')
```

```
# Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold (Biner)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(binary, cmap='gray')
plt.title("Biner")
plt.axis('off')
```

```
kernel = np.ones((5,5), np.uint8)

# Erosi
erosion = cv2.erode(binary, kernel, iterations=1)

# Dilatasi
dilation = cv2.dilate(binary, kernel, iterations=1)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(erosion, cmap='gray')
plt.title("Erosi")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(dilation, cmap='gray')
plt.title("Dilatasi")
plt.axis('off')
```

```
def mse(imageA, imageB):
    return np.mean((imageA - imageB) ** 2)

def psnr(imageA, imageB):
    mse_val = mse(imageA, imageB)
    if mse_val == 0:
        return 100
    return 20 * np.log10(255.0 / np.sqrt(mse_val))

def ssim_calc(imageA, imageB):
    return ssim(imageA, imageB)
```

```
print("=== Evaluasi Biner vs Asli ===")
print("MSE:", mse(gray, binary))
print("PSNR:", psnr(gray, binary))
print("SSIM:", ssim_calc(gray, binary))

print("\n=== Evaluasi Erosi vs Asli ===")
print("MSE:", mse(gray, erosion))
print("PSNR:", psnr(gray, erosion))
print("SSIM:", ssim_calc(gray, erosion))

print("\n=== Evaluasi Dilatasi vs Asli ===")
print("MSE:", mse(gray, dilation))
print("PSNR:", psnr(gray, dilation))
print("SSIM:", ssim_calc(gray, dilation))
```

```
titles = ['Asli', 'Biner', 'Erosi', 'Dilatasi']
images = [gray, binary, erosion, dilation]

plt.figure(figsize=(12,6))

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
```
